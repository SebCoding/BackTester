```python
# Initial imports
import numpy as np
import pandas as pd
from datetime import datetime
import datetime
import time
from pathlib import Path

%matplotlib inline
```

```python
## CoinGecko API: /coins/bitcoin/market_chart/range
#######################################################################
# Get historical market data include price, market cap, 
# and 24h volume within a range of timestamp (granularity auto)
#######################################################################
# Data granularity is automatic (cannot be adjusted)
# 1 day from query time = 5 minute interval data
# 1 - 90 days from query time = hourly data
# above 90 days from query time = daily data (00:00 UTC)

# datetime.datetime(year, month, day, hour, minute)
fromTimestamp = int(time.mktime(datetime.datetime(2021, 10, 1, 0, 0).timetuple()))
toTimestamp = int(time.mktime(datetime.datetime(2021, 11, 1, 0, 0).timetuple()))

# print(f"fromTimestamp = {fromTimestamp}")
# print(f"toTimestamp = {toTimestamp}")

url = f'https://api.coingecko.com/api/v3/coins/bitcoin/market_chart/range?vs_currency=usd&from={fromTimestamp}&to={toTimestamp}'
#print(url)
```

```python
# Query the API and get result in JSON format
import requests
import json

# Execute GET request and store response
#response_data = requests.get(url)

# Format data as a raw json file
#data = response_data.json()
#print(data)

# Add indents to JSON and output to screen
#print(json.dumps(data["prices"], indent=3))
```

```python
# Pandas read_json() accepts a URL directly, no need to convert JSON to DataFrame
df = pd.read_json(url)
#df.info()
df2 = df[["prices"]].copy()

df2[['UNIX Timestamp', 'Price']] = pd.DataFrame(df2.prices.tolist())
df2 = df2.drop('prices', 1)
#df2.info()

# Add column with readable datetime
#df2['Datetime'] = datetime.fromtimestamp(df2['UNIX Timestamp'])
df2['Datetime'] = pd.to_datetime(df2['UNIX Timestamp'], unit='ms')
Prices = df2[['Datetime', 'Price']]
Prices.set_index('Datetime', inplace=True)

# Remove milliseconds
Prices.index = Prices.index.floor('S').tz_localize(None)

Prices.tail(24)
```
